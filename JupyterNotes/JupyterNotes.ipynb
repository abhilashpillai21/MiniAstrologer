```python
import csv

document = []
metadata = []
ids = []
id =1
    
with open("../csv/menu_items.csv") as file:
    lines = csv.reader(file)
    
    for i, line in enumerate(lines):
        if(i==0):
            continue
        document.append(line[1])
        metadata.append({"item_id" : line[0]})
        ids.append(str(id))
        id+=1
```

```python
import chromadb
from chromadb.utils.embedding_functions import SentenceTransformerEmbeddingFunction

chroma_client = chromadb.Client()
```

```python
collection = chroma_client.create_collection(name="my_collection",
                                             embedding_function=SentenceTransformerEmbeddingFunction(
                                                 model_name="sentence-transformers/all-mpnet-base-v2"
                                                 )
                                            ) 
```

```python
collection.add(
    ids =ids,
    documents=document,
    metadatas = metadata
)
```

```python
results = collection.query(
    query_texts = ["donut"],
    n_results = 5
)
results["documents"][0]
results["distances"][0]
top_chunks = list(zip(results["distances"][0], results["documents"][0]))
context = "\n\n".join(f"""Source {i+1} : score:{score:.4f}\n{text}""" for i, (score, text) in enumerate(top_chunks))
context
```

```python
chroma_client.delete_collection(name="my_collection")
```
